```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Virat_ODI.csv',encoding='latin1')
df.info()
df.head()
```

```python
df2=df.drop(['Date','mins','NotOut','bf'],axis='columns')
df2.head(3)
```

```python
dumm = pd.get_dummies(df2.Opp)
dumm.head(3)
```

```python
dumm1 = pd.get_dummies(df2.Ground)
dumm1.head(3)
```

```python
df3=pd.concat([df2,dumm,dumm1],axis='columns')
df4=df3.drop(['Ground','Opp'],axis='columns')
```

```python
x=df4.drop(['runs','Inns'],axis='columns')
y=df4.runs
x.head(3)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_test.head(3)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
